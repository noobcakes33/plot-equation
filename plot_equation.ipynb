```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

# constant range for the x axis
t = np.arange(1, 10, 0.1)

# reading the csv file
df = pd.read_csv('c_values.csv')

# reading the c values from C_values column
c = df['C_values']

# creating a figure
fig = plt.figure()

for value in c:
    # loop over all values of the csv file
    # and make a plot for each value of c
    # following this equation
    p = value / t
    plt.plot(t, p)
    plt.legend(c)

# P for y axis
plt.ylabel('P')
# t for x axis
plt.xlabel('t')

# showing the plot
plt.show()
fig.savefig('equation_plotted.png')
```
